```python
import numpy as np

x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)

m = 0
b = 0
learning_rate = 0.01

for i in range(1000):
    y_pred = m * x + b
    dm = (-2/len(x)) * sum(x * (y - y_pred))
    db = (-2/len(x)) * sum(y - y_pred)

    m = m - learning_rate * dm
    b = b - learning_rate * db

print("Slope =", round(m, 2))
print("Intercept =", round(b, 2))
```
